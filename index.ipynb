```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('C:\\pythonproject3\\products.xlsx')
print(df)
```

```python
df50=df[90:101]

plt.plot(df50["Price"],df50["Stock"])
plt.xlabel("Price")
plt.ylabel("Stock")
plt.title("Price VS Stock")
plt.show()
```

```python
plt.bar(df50["Name"],df50["Price"])
plt.xticks(rotation=90)
plt.ylabel("price")
plt.title("product price comparison")
plt.show()
```

```python

availability_counts = df['Availability'].value_counts()

plt.pie(
    availability_counts.values,
    labels=availability_counts.index,
    autopct='%1.1f%%'
)

plt.title("Availability Distribution")
plt.show()
```

```python
plt.scatter(df50["Stock"], df50["Price"])
plt.xlabel("Stock")
plt.ylabel("Price")
plt.title("Stock vs Price Scatter")
plt.show()
```

```python
plt.hist(df50["Price"], bins=10)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Price Histogram")
plt.show()
```

```python
plt.boxplot(df50["Price"])
plt.title("Price Boxplot")
plt.ylabel("Price")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('C:\\pythonproject3\\products.xlsx')
print(df)
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

corr = numeric_df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
data = df[["Price", "Stock"]]

sns.pairplot(data)
plt.show()
```

```python
sns.boxplot(x="Category", y="Stock", data=df)
plt.xticks(rotation=90)
plt.title("Stock Distribution by Category")
plt.show()
```

```python
df50=df[50:61]
sns.countplot(x="Brand", data=df50)
plt.xticks(rotation=90)
plt.title("Count of Products by Brand")
plt.show()
```
